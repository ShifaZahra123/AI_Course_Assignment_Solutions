Write a function called calculate_sum that takes a list of numbers as an
argument and returns the sum of those numbers.
o Example: Input: calculate_sum([1, 2, 3, 4, 5]) Output: 15

```python
def calculate_sum(numbers):
  sum = 0 # initialize sum as zero to avoid garbage value
  for num in numbers:
    sum += num  # every num from list is stored in sum iteratively
  return sum

numbers = [1, 2, 3, 4, 5]
result = calculate_sum(numbers)
print(result)  # Output: 15
```

Write a function called remove_duplicates that takes a list as an argument and
returns a new list with duplicate elements removed.

Input: remove_duplicates([1, 2, 2, 3, 4, 4, 5]) Output: [1, 2, 3, 4, 5]

```python
def remove_duplicates(numbers):
  return list(set(numbers)) # Set donot allow duplicate members
# convert set into list, because output required in list format
#  set function is a built-in function that converts an iterable (like a list, tuple, or string) into a set.
# list function is a built-in function that converts an iterable (like a list, tuple, or string) into a set.
numbers = [1, 2, 2, 3, 4, 4, 5]
result = remove_duplicates(numbers)
print(result)  # Output: [1, 2, 3, 4, 5]
```

Write a function called find_max that takes a list of numbers as an argument and
returns the largest number in the list.
o Example: Input: find_max([10, 5, 8, 20, 3]) Output: 20

```python
def find_max(numbers): # numbers is a parameter here
  return max(numbers) # max is built-in function that returns largest element from list

num = [10, 5, 8, 20, 3]
result = find_max(num) # num is argument here
print(result)  # Output: 20
```
