Write a program that prints the multiplication table of a given number using a for
loop.

```python
# In Python , user input taken as string, so must be convert in int data type
x = int(input("Enter number you print table for that number: "))
rang = range(1,11) # Range from 1st index to last index-1 (n-1)
for table in  rang:
   res= x*table # Multiplication
   print(x,"*",table,"=",res)
```

Write a program that prints all the even numbers from 1 to 10 using a while loop

```python
i = 1
while i <= 10:
    if i % 2 == 0: # Check if num is multiple of 2, it means it is even number
      print("Number is even: ",i)
    i = i + 1
```

Write a function called "greet_user" that takes a user's name as an argument and
prints a personalized greeting message.

```python
# Function definition with a parameter
def greet_user(name): # Here, name is parameter
  print("Hello, " , name , "!")

# Function calling with an argument
name = input("Enter your name: ")
greet_user(name) # Here, name is argument
```

Write a function called "is_even" that takes a number as an argument and returns
True if the number is even, and False otherwise.

```python
def is_even(num): # num is parameter
  if num % 2 == 0: # If num is multiple of 2, it is even otherwise odd
    return True
  else:
    return False
# In Python , user input taken as string, so must be convert in int data type
number = int(input("Enter a number: "))
if is_even(number): # number is argument
  print(number, "is an Even Number.")  # print if returns True
else:
  print(number, "is an Odd Number.") # execute if returns False
```
