```python
# Remove rows with missing data
full_data.dropna(inplace = True)
```

```python
full_data.shape
```

```python
full_data.head()
```

```python
# Convert objects to category data type
objcat = ['ocean_proximity']

for colname in objcat:
    full_data[colname] = full_data[colname].astype('category')
```

```python
# Identify categorical features
full_data.select_dtypes(['category']).columns
```

```python
# Convert categorical variables into 'dummy' or indicator variables
ocean_proximity = pd.get_dummies(full_data['ocean_proximity'], drop_first = True) # drop_first prevents multi-collinearity
```

```python
# Add new dummy columns to data frame
full_data = pd.concat([full_data, ocean_proximity], axis = 1)
full_data.head(5)
```

```python
# Drop unecessary columns
full_data.drop(['ocean_proximity'], axis = 1, inplace = True)

# Shape of train data
print('train_data shape',full_data.shape)

# Confirm changes
full_data.head()
```

## **Target Variable Splitting**
We will spilt the Full dataset into **Input** and **target** variables

**Reason for selecting Target Variable (median_house_value):** The goal of the analysis is to predict house prices

```python
# Split data to be used in the models
# Create matrix of features
x = full_data.drop('median_house_value', axis = 1) # grabs everything else but 'median_house_value'

# Create target variable
y = full_data['median_house_value'] # y is the column we're trying to predict
```

```python
# x Represents the Features
x.shape
```

```python
# y represents the Target
y.shape
```

```python
# Use x and y variables to split the training data into train and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .20, random_state = 101)
```

```python
x_train.shape
x_train
```

```python
# y_train.shape
y_train
```

```python
x_test.shape
y_test.shape
x_test
```

### **LOGISTIC REGRESSION**

## **Model Training**

```python
# Fit
# Import model
from sklearn.linear_model import LogisticRegression

# Create instance of model
lreg = LogisticRegression()

# Pass training data into model
lreg.fit(x_train, y_train)
```

## **Model Testing**

## **Class prediction**

```python
# Predict
y_pred_lreg = lreg.predict(x_test)
print(y_pred_lreg)
print(y_pred_lreg.shape)
```

```python
probabilities = lreg.predict_proba(x_test)[:, 1]
print(probabilities.shape)
print(probabilities)
```

## **Probabilities of each class**

```python
for i in range(len(y_test)):
  print(round(probabilities[i], 3)," --> ", y_pred_lreg[i])
```

## **Model Evaluation**

```python
# Score It
from sklearn.metrics import classification_report, accuracy_score

print('Classification Model')
# Accuracy
print('--'*30)
logreg_accuracy = round(accuracy_score(y_test, y_pred_lreg) * 100,2)
print('Accuracy', logreg_accuracy,'%')
```
