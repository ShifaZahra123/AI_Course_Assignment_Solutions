# **Task 3: Descriptive Statistics**

```python
df.describe()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Mean
mean_values = numeric_df.mean()

# Median
median_values = numeric_df.median()

# Minimum
min_values = numeric_df.min()

# Maximum
max_values = numeric_df.max()

# Quantiles (for example: 25%, 50%, and 75%)
quantiles = numeric_df.quantile([0.25, 0.50, 0.75])

# Create a DataFrame to hold the summary statistics
summary_stats = pd.DataFrame({
    'mean': mean_values,
    'median': median_values,
    'min': min_values,
    'max': max_values,
    '25%': numeric_df.quantile(0.25),
    '50%': numeric_df.quantile(0.50),
    '75%': numeric_df.quantile(0.75)
})

# Display the summary statistics
print(summary_stats)
```
